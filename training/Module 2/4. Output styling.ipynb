```python
from ipywidgets import widgets, Layout, HTML
from IPython.display import display
```

```python
border = '4px dashed red'
#border = 'none'

out = widgets.Output(layout=Layout(width='99%', min_width='600px', height='600px', border=border))
out.add_class('box')
out
```

```python
out.clear_output()
with out:
    display(HTML('''
<style>
.box {
    background-color: lightgreen;
}
</style>
'''))
```

```python
out.clear_output()
with out:
    display(HTML('''
<style>
.box {
    background: repeating-linear-gradient(45deg,  #606dbc,  #606dbc 10px,  #465298 10px,  #465298 20px);
}
</style>
'''))
```

```python
out.layout.border = None

out.clear_output()
with out:
    display(HTML('''
<style>
.box {
    background-image: linear-gradient(to right, rgba(255,0,0,0), rgba(255,0,0,1));
}
</style>
'''))    
```

```python
# Display an image inside the output widget
out.layout.border = '4px dashed red'

from urllib.request import urlopen
from PIL import Image

imagestr = ''
img = Image.open(urlopen(imagestr))


out.clear_output()
with out:
    display(img)
```

```python
# Fill the output with an image in base64 format (seamless pattern)
out.layout.border = None

out.clear_output()
with out:
    display(HTML('''
<style>
.box {
    background-image: url('');
}
</style>
'''))    
```

```python
# Fill the output with an image from an URL
out.layout.border = None

out.clear_output()
with out:
    display(HTML('''
<style>
.box {
    background-image: url('https://picsum.photos/id/1050/600/400?grayscale');
}
</style>
'''))    
```
