```python
# FULLSCREEN EXAMPLE WITH BACKGROUND IMAGE AN RESPONSIVE PANELS
import ipyvuetify as v
import numpy as np
import pandas as pd
import plotly.express as px
from ipywidgets import widgets, Layout, HTML

from vois.vuetify import app, switch, datePicker
from vois import svgHeatmap, svgBubblesChart

%run app.py

# Click on a tab of the title
def on_click_tab(arg):
    g_app.snackbar(arg)

# Click on the credits text
def on_click_credits():
    g_app.snackbar('CREDITS')
    
# Click on the BDAP text
def on_click_credits2():
    g_app.urlOpen('https://jeodpp.jrc.ec.europa.eu/bdap/')
    
# Click on the logo
def on_click_logo():
    g_app.snackbar('LOGO')

# Click on the footer buttons
def on_click_footer(arg):
    g_app.snackbar(arg)
    
# Click on the footer minipanel
def on_click_minipanel(index):
    g_app.snackbar(str(index))

    
# Example of a text with shadow
title = 'Background image and panels'
titlesvg = '''
<svg width="550" height="75" viewBox="0 0 550 75">
  <g style="overflow:hidden; text-anchor: middle; font-size:28; font-weight: bold; font-family: Arial, Helvetica, sans-serif">
    <defs>
      <filter id="shadow" x="-20%%" y="-20%%" width="140%%" height="140%%">
        <feGaussianBlur stdDeviation="2 2" result="shadow"/>
        <feOffset dx="6" dy="6"/>
      </filter>
    </defs>
    <text style="filter: url(#shadow); fill: grey" x="275" y="50">%s</text>
    <text x="275" y="50" style="fill: black">%s</text>
  </g>
</svg>
''' % (title, title)

g_app = app(title='',
                titlesvg=titlesvg,
                titleheight='6vh',
                titlecredits='Created by Unit I.3',
                titlecredits2='BDAP',
                titlewidth='42vw',
                titlestyle='color: black; font-weight: bold; font-size: 26px;',
                titlespacestyle='width: 10vw;',
                dark=True,
                backgroundimageurl='',
                titletabs=['Charts', 'Tables', 'Maps'],
                titletabsdark=True,
                sidepanelwidth=0,
                sidepaneltext="Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua. Ut enim ad minim veniam, quis nostrud exercitation ullamco laboris nisi ut aliquip ex ea commodo consequat. Duis aute irure dolor in reprehenderit in voluptate velit esse cillum dolore eu fugiat nulla pariatur. Excepteur sint occaecat cupidatat non proident, sunt in culpa qui officia deserunt mollit anim id est laborum.",
                sidepanelcontent=[v.Icon(class_='pa-0 ma-0 ml-2', children=['mdi-help'])],
                sidepaneldark=True,
                sidepanelbackdark=False,
                minipanelicons=['fa-truck', 'mdi-heart', 'mdi-magnify'],
                minipaneltooltips=['Function 1', 'Function 2', 'Function 3'],
                minipanellarge=False,
                minipanelopen=False,
                minipanelbuttoncolor='black',
                minipaneliconscolor='#010101',
                footercolor='#888888',
                footerdark=True,
                footerbuttons=[],
                footerheight='4vh',
                onclickminipanel=on_click_minipanel,
                onclicktab=on_click_tab,
                onclickcredits=on_click_credits,
                onclickcredits2=on_click_credits2,
                onclicklogo=on_click_logo,
                onclickfooter=on_click_footer,
                fullscreen=True)


# Vertically,   from top to bottom: 6vh(titlebar) + 2vh + 40vh(panel) + 2vh + 42vh(bigpanel) + 2vh + 4vh(footerbar)
# Horizontally, from left to right: 3vw + 46vw + 2vw + 46vw + 3vw
#                                   3vw + 94vw + 3vw

# Dimension of the small panels
widthA  = '46vw'
heightA = '40vh'

# Dimensions of the big panel
widthB  = '94vw'
heightB = '42vh'

# Positioning of the panels (from the top'left of the app.outcontent)
left1 = '3vw'
top1  = '2vh'

left2 = '51vw'
top2  = '2vh'

left3 = '3vw'
top3  = '44vh'

#g_app.contentBackground('https://picsum.photos/id/1050/2800/1600?grayscale')
g_app.contentBackground('')

g_app.contentResetPanels()
out1 = g_app.contentAddPanel(widthA, heightA, left1, top1, '1px solid grey', '#ffffff88')  # Transparent
out2 = g_app.contentAddPanel(widthA, heightA, left2, top2, '1px solid grey', '#ffffff88')
out3 = g_app.contentAddPanel(widthB, heightB, left3, top3, '1px solid grey', '#ffffff00')

s = switch.switch(True, "Sample switch", inset=True)
datemin = datePicker.datePicker(date='2022-09-12', label='Min date:', offset_x=False, offset_y=True)

space = v.Html(tag='div', children=[''], class_="pa-0 ma-0", style_='width: 10px; height: 10px;')
with out1:
    #display(s.draw())
    df = pd.DataFrame(np.random.random((20,50)))
    colorlist = ['#ffffff', '#f4ae1a']
    display(svgHeatmap.heatmapChart(df, 
                                    title='Mushrooms production',
                                    textRows='Days',
                                    textColumns='Quantity',
                                    wTitle=7.0,
                                    hTitle=5.5,
                                    width=45.0,
                                    height=38.0,
                                    colorlist=colorlist,
                                    backcolor='#ffffff00'))
    
with out2:
    #display(widgets.HBox([space,datemin.draw()]))
    colorlist = ['#f4a70d', '#f56b0f', '#f53a0f', '#f5e70f', '#c2f50f', '#3cfccd', '#999999']

    df = pd.read_csv('https://jeodpp.jrc.ec.europa.eu/services/shared/csv/mushrooms.csv')

    xcolumn     = 'cap-shape'
    ycolumn     = 'cap-surface'
    colorcolumn = 'habitat'
    sizecolumn  = 'count'
    dfgrouped = df.groupby([xcolumn, ycolumn, colorcolumn]).size().reset_index(name=sizecolumn)

    dfgrouped.columns = ['Shape', 'Cap surface', 'Mushroom habitat:', 'count']

    b = svgBubblesChart.svgBubblesChart(dfgrouped,
                                        width=45.0,
                                        height=39.0,
                                        xcolumn=dfgrouped.columns[0],
                                        ycolumn=dfgrouped.columns[1],
                                        colorcolumn=dfgrouped.columns[2],
                                        sizecolumn=dfgrouped.columns[3],
                                        strokewidth=1,
                                        strokecolor='#ffffff',
                                        backcolor='#f0f0f0',
                                        backlinecolor='#000000',
                                        bubblecolors=colorlist,
                                        fontsize=1.1,
                                        title='Mushrooms analysis',
                                        mode='spread', legendrows=1, legenditemwidth=5.0)
    display(b.draw())

with out3:
    df = px.data.tips()
    fig = px.histogram(df, x="total_bill", color="sex", marginal="rug", template='plotly_white',
                       color_discrete_sequence=['#f4ae1a', '#999999'])
    fig.update_layout(title='<b>Mushrooms delivery per day and sex</b>',
                      xaxis=dict(title='Days', linecolor='#888888', gridcolor='#dddddd'),
                      yaxis=dict(linecolor='#888888', gridcolor='#dddddd'),
                      autosize=True,
                      titlefont=dict(family="Roboto",size=24,color='#000000'))
    fig.show()    


g_app.show()
```
