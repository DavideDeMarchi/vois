```python
import geopandas as gpd
import json
from ipyleaflet import Map
from ipywidgets import widgets

import dataread

debug = widgets.Output()
display(debug)

%run geojsonLayer.py


def on_click(feature):
    with debug:
        if feature is None:
            print('on_click:', None)
        else:
            print('on_click:', feature['properties']['City_code'])

def featureColor(feature):
    return {'radius': 9, 'fillColor': "#ff0000", 'color': "#000000", 'weight': 1.5, 'opacity': 1.0, 'fillOpacity': 0.8}
    
m = Map(center=[50,12], zoom=5, scroll_wheel_zoom=True)

df = dataread.read()
dfcities = df.groupby(by=['City_code'], dropna=False).first().reset_index()[['City_code', 'Lat', 'Lon']]
gdf = gpd.GeoDataFrame(dfcities, geometry=gpd.points_from_xy(dfcities.Lon, dfcities.Lat))

feature_collection = json.loads(gdf.to_json())
layer = geojsonLayer(feature_collection,
                     on_click=on_click,
                     style_callback=featureColor)

layer.addToMap(m)
m
```
